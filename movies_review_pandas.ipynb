The BEST and WORST movie

```python
from IPython.display import HTML
```

```python
df_best = df [['poster_path','title','budget_musd','revenue_musd','vote_count','vote_average','popularity']].copy()
```

```python
df_best
```

```python
df_best['profit_musd'] = df.revenue_musd.sub(df.budget_musd)
df_best['return'] = df.revenue_musd.div(df.budget_musd)
```

```python
df_best.head()
```

**change column names for display purposes**

```python
df_best.columns=['','Title','Budget','Revenue','Votes', 'Average Rating', 'Popularity','Profit','ROI']
```

```python
df_best.set_index('Title', inplace=True)
```

```python
df_best.iloc[0,0]
```

```python
subset = df_best.iloc[:5, :2]
subset
```

```python
HTML(subset.to_html(escape=False))
```

```python
df_best.sort_values(by='Average Rating', ascending=False)
```

```python
df_best.sort_values(by='ROI', ascending=False)
```

```python
df_best.loc[df_best.Budget>5].sort_values(by='ROI', ascending=False)
```

```python
df_best.Budget.fillna(0, inplace=True)
df_best.ROI.fillna(0, inplace=True)
```

```python
df_best.loc[df_best.Budget>5].sort_values(by='ROI', ascending=False)
```

```python
def best_worst(n, by , ascending=False, min_bud=0, min_votes=0):
  df2= df_best.loc[(df_best.Budget >= min_bud) & 
                   (df_best.Votes >= min_votes),
                   ["", by]].sort_values(by = by, ascending=ascending).head(n).copy()
  return HTML(df2.to_html(escape=False))
```

**Best Movies/worst movies**

```python
best_worst(n= 5, by ="Revenue")
```

**Best Movies By Budget**

```python
best_worst(n= 5, by ="Budget")
```

**Movies lowest profit**

```python
best_worst(n= 5, by ="Profit", ascending=True)
```

**Movies Top ROI**

```python
best_worst(n= 5, by ="ROI", min_bud=5)
```

**Movies lowest ROI**

```python
best_worst(n= 5, by ="ROI", ascending=False, min_bud=50)
```

```python
#top 5 by Rating
best_worst(n= 5, by ="Average Rating", min_votes=10)
```

```python
best_worst(n= 5, by ="Average Rating", min_votes=10 , min_bud=5)
```

```python
#most popular
best_worst(n= 5, by ="Votes", min_votes=10 , min_bud=5)
```
